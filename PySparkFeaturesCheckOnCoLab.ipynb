### derive colum for year 2019, 2020 with category wise total sales from dataset we have different year sales with category

```python
dataSample = [
        ("1", "Office Supplies", "2020-09-02", 65.94),
        ("2", "Furniture", "2021-12-04", 11.35),
        ("3", "Office Supplies", "2019-06-12", 24.56),
        ("4", "Office Supplies", "2020-09-04", 16.96),
        ("5", "Technology", "2020-09-04", 279.95),
        ("6", "Office Supplies", "2018-04-21", 828.84),
        ("7", "Office Supplies", "2019-11-05", 7.21),
        ("8", "Office Supplies", "2019-11-05", 27.2),
        ("9", "Technology", "2018-11-03", 89.97),
        ("10", "Furniture", "2020-02-16", 318.08),
        ]
dataFrameTest = spark.createDataFrame(dataSample, ["ID", "Category", "OrderDate", "Sales"])
dataFrameTest.show()
```

```python
from pyspark.sql.functions import *

dataFrameTest \
  .withColumn('Year',  year('OrderDate')) \
  .select('Category', 'Year').show()
```

```python
dataFrameTest \
    .filter( year('OrderDate') == 2019 ) \
    .show()
```

```python
 # Calculate Year Wise Sales

 # .filter( year(dataFrameTest.OrderDate).isin (2019, 2020)) \

dataFrameTest \
    .where( year(dataFrameTest.OrderDate).isin (2019, 2020)) \
    .groupBy('Category') \
    .agg(
         sum(
             when (
                 year('OrderDate') == 2019,
                 col('Sales')
                 ) \
              .otherwise(
                  0)
              ) \
          .alias('Sales2019'),
         sum(
             when (
                   year('OrderDate') == 2020,
                   dataFrameTest.Sales
                   ) \
              .otherwise(
                  0)
              ) \
          .alias('Sales2020')
          ) \
    .select(col('Category'), 'Sales2019', 'Sales2020') \
    .orderBy(col('Category').desc()) \
    .show()
```

### When Conditions with And or Or

```python
dataFrame \
.withColumn(
    "TransactionSize",
    when(
        (col('dlbytes') > 500000) & (col('ulbytes') > 500000),
        'Large'
    )
    .otherwise(
        'Small')
).show()
```

```python
dataFrame \
.withColumn(
    "TransactionSize",
    when(
        (col('dlbytes') > 500000) & (col('ulbytes') > 500000),
        'Large')
    .when(
        (col('dlbytes') > 250000) & (col('ulbytes') > 250000),
        'Medium')
    .when(
        (col('dlbytes') > 150000) & (col('ulbytes') > 150000),
        'Small')
    .otherwise(
        'Tiny')
) \
.show()
```

```python
dataFrame \
.withColumn(
    'Status',
    when(col('txn_time') * 1000 > 2000,
         'Active')
    .otherwise(
        'Inactive')
    ).show()
```

### Window Function - Rank()

> Indented block



```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

dataFrame.withColumn(
    'Rank',
    rank().over(
        Window.partitionBy('Country', 'platform')
        .orderBy(col('dlbytes').desc())
    )
).select('country', 'platform', 'dlbytes', 'Rank') \
.show(6000)
```

### Window Function - Row_Number()

```python
dataFrame.withColumn(
    'RowNumber',
    row_number().over(
        Window.partitionBy('Country', 'platform')
        .orderBy('date_time')
        )
    ) \
.select('Country', 'platform', 'date_time', 'RowNumber') \
.show(6000)
```

### Window Functions - Lead/Lag

```python
dataFrame \
  .withColumn (
      'PreviousBytes',
      lag('dlbytes').over (
          Window.partitionBy()
          .orderBy('date_time')
      )
  ) \
  .withColumn (
      'NextBytes',
      lead('dlbytes').over (
          Window.partitionBy()
          .orderBy('date_time')
      )
  ) \
.select('date_time', 'dlbytes', 'PreviousBytes', 'NextBytes') \
.show()
```
